```python
#Inicio de overbooking

import pandas as pd

# Lendo os CSVs
df_dia = pd.read_csv('f_diarias.csv')
df_inv = pd.read_csv('f_inventario_diario.csv')

# Fazendo o merge pelas 3 colunas em comum
df_dia_e_inv = pd.merge(
    df_dia,
    df_inv,
    on=['hotel_id', 'data_id', 'tipo_quarto'],
    how='inner'  # tipo de junção
)

df_dia_e_inv


```

```python
#2. Tratar as diárias e calcular se existe overbooking

#Preciso que o número de quartos disponíveis no hotal seja maior o igual ao número de quartos vendidos para que não acha overbooking

df_total = df_dia_e_inv['quartos_disponiveis'] >= df_dia_e_inv['quartos_vendidos']

df_total
```

```python
(df_dia_e_inv['quartos_disponiveis'] >= df_dia_e_inv['quartos_vendidos']).all()
```

```python
#Ver os resultados falsos para achar os overbookings:


df_overbooking = df_dia_e_inv[df_dia_e_inv['quartos_disponiveis'] < df_dia_e_inv['quartos_vendidos']]

df_overbooking
```

```python
#3. Devolver tabela tratada sem overbooking

df_trat = df_dia_e_inv[df_dia_e_inv['quartos_disponiveis'] >= df_dia_e_inv['quartos_vendidos']].reset_index(drop=True)

df_trat


```

```python
df_overbooking.to_csv("Overbooking.csv", sep=';', index=False)
df_overbooking
```

```python
#3- Função para transformar números (YYYYMMDD) em datas reais
def formatar_data_humana(valor):
    try:
        return pd.to_datetime(str(int(valor)), format='%Y%m%d')
    except:
        return pd.NaT
```

```python
#GRÁFICOS:

#Antecedência da Reserva
plt.figure()
sns.histplot(df_res_trat['lead_time'], bins=30, kde=True, color='teal')

plt.xlim(0, 150)  # exemplo: de 0 a 150 dias

plt.xticks(range(0, 150, 5))  # escala visual, de 10 e, 10 dias


plt.title('Hábito do Cliente: Quantos dias antes eles reservam?')
plt.xlabel('Dias de Antecedência')
plt.ylabel('Quantidade de Reservas')



dados = df_res_trat['lead_time']

contagens, bins = np.histogram(dados, bins=30)
indice_max = contagens.argmax()
x_max = (bins[indice_max] + bins[indice_max + 1]) / 2
y_max = contagens[indice_max]


# ponto máximo
plt.scatter(x_max, y_max, color='violet', zorder=5)
plt.annotate(
    f'Máx: {int(x_max)} dias',
    xy=(x_max, y_max),
    xytext=(x_max + 10, y_max),
    arrowprops=dict(arrowstyle='->', color='violet')
)

plt.title('Hábito do Cliente: Quantos dias antes eles reservam?')
plt.xlabel('Dias de Antecedência')
plt.ylabel('Quantidade de Reservas')

plt.savefig('grafico_1_leadtime.png')
plt.show()



plt.savefig('grafico_1_leadtime.png')
```
