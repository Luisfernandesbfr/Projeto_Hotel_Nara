```python
#Inicio de overbooking

import pandas as pd

# Lendo os CSVs
df_dia = pd.read_csv('f_diarias.csv')
df_inv = pd.read_csv('f_inventario_diario.csv')

# Fazendo o merge pelas 3 colunas em comum
df_dia_e_inv = pd.merge(
    df_dia,
    df_inv,
    on=['hotel_id', 'data_id', 'tipo_quarto'],
    how='inner'  # tipo de junção
)

df_dia_e_inv


```

```python
#2. Tratar as diárias e calcular se existe overbooking

#Preciso que o número de quartos disponíveis no hotal seja maior o igual ao número de quartos vendidos para que não acha overbooking

df_total = df_dia_e_inv['quartos_disponiveis'] >= df_dia_e_inv['quartos_vendidos']

df_total
```

```python
(df_dia_e_inv['quartos_disponiveis'] >= df_dia_e_inv['quartos_vendidos']).all()
```

```python
#Ver os resultados falsos para achar os overbookings:


df_overbooking = df_dia_e_inv[df_dia_e_inv['quartos_disponiveis'] < df_dia_e_inv['quartos_vendidos']]

df_overbooking
```

```python
#3. Devolver tabela tratada sem overbooking

df_trat = df_dia_e_inv[df_dia_e_inv['quartos_disponiveis'] >= df_dia_e_inv['quartos_vendidos']].reset_index(drop=True)

df_trat


```

```python
df_overbooking.to_csv("Overbooking.csv", sep=';', index=False)
df_overbooking
```

```python
df_trat.to_csv("f_diarias_e_inventario_trat.csv", sep=';', index=False)
df_trat
```
